```python
df1 = pd.read_csv("BN_CV_more_alleles.csv")
df2 = pd.read_csv("BN_CV_other_alleles.csv")
df = pd.concat([df1, df2])

df["combined-score"] = 2*(df["auc"]-0.5) + df["f1"]
```
