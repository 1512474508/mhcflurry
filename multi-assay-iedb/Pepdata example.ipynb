```python
df = pepdata.iedb.tcell.load_dataframe()
```

```python
df_hiv1_mask = df["Antigen Source Organism Name"] == "Human immunodeficiency virus 1"
```

```python
df_hiv1 = df[df_hiv1_mask]
df_hiv1.iloc[0]
```
