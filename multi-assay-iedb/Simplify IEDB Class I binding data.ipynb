```python
###
# Create a DataFrame with simplified columns and only peptidic entries 
# (no conformational epitopes or whole protein antigens)
###


from collections import OrderedDict

def make_simplified_dataframe(df):
    
    df_peptides =  df[df["Epitope"]["Object Type"] == "Linear peptide"]


    columns = OrderedDict()
    columns["allele"] = df_peptides["MHC"]["Allele Name"]
    columns["mhc_class"] = df_peptides["MHC"]["MHC allele class"]
    columns["peptide"] = df_peptides["Epitope"]["Description"]
    columns["peptide_length"] = df_peptides["Epitope"]["Description"].str.len()
    columns["organism"] = df_peptides["Epitope"]["Organism Name"]
    columns["antigen_name"] = df_peptides["Epitope"]["Antigen Name"]
    columns["assay_group"] = df_peptides["Assay"]["Assay Group"]
    columns["assay_method"] = df_peptides["Assay"]["Method/Technique"]
    columns["assay_name"] = (
        df_peptides["Assay"]["Assay Group"] + " ::: " + 
        df_peptides["Assay"]["Method/Technique"])
    columns["assay_id"] = df_peptides["Assay"]["Assay Type ID"]

    columns["units"] = df_peptides["Assay"]["Units"]
    meas = np.array(df_peptides["Assay"]["Quantitative measurement"])
    nan_mask = np.isnan(meas)
    qual = df_peptides["Assay"]["Qualitative Measure"]
    meas[nan_mask] = qual[nan_mask].str.startswith("Positive")
    columns["meas"] = meas
    columns["is_quantitative"] = ~nan_mask
    columns["qualitative_meas"] = qual
    columns["reference_authors"] = df_peptides["Reference"]["Authors"]
    
    columns["reference_journal"] = df_peptides["Reference"]["Journal"]
    columns["reference_title"] = df_peptides["Reference"]["Title"]
    columns["reference_date"] = df_peptides["Reference"]["Date"]
    
    return pd.DataFrame(columns)

df_simplified = make_simplified_dataframe(df)
df_simplified.ix[0]
```

```python
df_simplified.to_csv("iedb-condensed.csv")
```
